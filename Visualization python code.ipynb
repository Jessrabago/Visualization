```python
# Convert the column names into the index
Total_numberSites= Total_numberSites.transpose()

Total_medianCost=Total_medianCost.transpose()

Total_meanCost=Total_meanCost.transpose()
Total_meanCost

```

```python
#rename the column
Total_numberSites = Total_numberSites.rename(columns={12: 'Number of sites'})
Total_medianCost=Total_medianCost.rename(columns={12: 'Median Cost'})
Total_meanCost=Total_meanCost.rename(columns={12: 'Mean Cost'})
Total_meanCost
```

```python
#concat the colums for all the dataframes
Total_summary_0_4kw = pd.concat([Total_numberSites, Total_medianCost, Total_meanCost], axis=1)
Total_summary_0_4kw
```

```python
# Create a figure and an axes object
fig, ax = plt.subplots()



# Define the update function
def update(num):
    # Clear the axes
    ax.clear()
    # Select the data for the current frame
    data = Total_summary_0_4kw.iloc[num, :]
    # Plot the bar chart
    ax.bar(data.index, data.values)
    # Add a text label with the index value
    ax.text(0.5, 0.9, Total_summary_0_4kw.index[num], transform=ax.transAxes, fontsize=20)
    

# Create the animation object
ani = FuncAnimation(fig, update, frames=range(9), repeat=True)

# Display the animation inline in the notebook
display(HTML(ani.to_jshtml()))
```
